```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from scipy.optimize import curve_fit
from scipy.stats import poisson
from scipy.stats import pearsonr
```

```python
BA1 = nx.barabasi_albert_graph(1000,1,seed=19)
BA2 = nx.barabasi_albert_graph(1000,2,seed=19)
BA3 = nx.barabasi_albert_graph(1000,3,seed=19)
#nx.draw(BA3,with_labels=True,node_color='red')
```

```python
def plot_degree_dist_log_bin(G1,G2,G3):

    degrees1 = [G1.degree(n) for n in G1.nodes()]
    degrees2 = [G2.degree(n) for n in G2.nodes()]
    degrees3 = [G3.degree(n) for n in G3.nodes()]
    h1,bin1 = np.histogram(degrees1,bins=np.logspace(np.log10(0.1),np.log10(max(degrees1)), 10), density = True)
    h2,bin2 = np.histogram(degrees2,bins=np.logspace(np.log10(0.1),np.log10(max(degrees2)), 10), density = True)
    h3,bin3 = np.histogram(degrees3,bins=np.logspace(np.log10(0.1),np.log10(max(degrees3)), 10), density = True)
    plt.loglog(bin1[1:],h1)
    plt.loglog(bin2[1:],h2)
    plt.loglog(bin3[1:],h3)
    plt.loglog(bin3[1:],1/bin3[1:]**3)
    plt.show()
plot_degree_dist_log_bin(BA1, BA2, BA3)
```

```python
def plot_degree_dist_log_bin(G1,G2,G3):
    degrees1 = sorted([G1.degree(n) for n in G1.nodes()])
    degrees2 = sorted([G2.degree(n) for n in G2.nodes()])
    degrees3 = sorted([G3.degree(n) for n in G3.nodes()])
    cumDeg1 = np.cumsum(degrees1)[::-1]
    cumDeg2 = np.cumsum(degrees2)[::-1]
    cumDeg3 = np.cumsum(degrees3)[::-1]
    h1,bin1 = np.histogram(cumDeg1,bins=np.logspace(np.log10(0.1),np.log10(max(cumDeg1)), 10), density = True)
    h2,bin2 = np.histogram(cumDeg2,bins=np.logspace(np.log10(0.1),np.log10(max(cumDeg2)), 10), density = True)
    h3,bin3 = np.histogram(cumDeg3,bins=np.logspace(np.log10(0.1),np.log10(max(cumDeg3)), 10), density = True)
    plt.loglog(bin1[1:],h1)
    plt.loglog(bin2[1:],h2)
    plt.loglog(bin3[1:],h3)
    plt.loglog(bin3[1:],1/bin3[1:]**3)
    plt.show()
plot_degree_dist_log_bin(BA1, BA2, BA3)
```

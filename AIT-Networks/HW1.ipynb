```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from scipy.optimize import curve_fit
from scipy.stats import poisson
from scipy.stats import pearsonr
```

```python
# Fully connected graph
FCG = nx.Graph()
FCG.add_edge("P","S")
FCG.add_edge("R","S")
FCG.add_edge("Q","P")
FCG.add_edge("Q","R")
nx.draw(FCG,with_labels=True)
```

```python
BA = nx.barabasi_albert_graph(50,2,seed=19)
for eachNode in BA.nodes():
    for eachNextNode in BA.nodes():
        if eachNode == eachNextNode:
            continue
        BA.add_edge(eachNode,eachNextNode)
nx.draw(BA,with_labels=True,node_color='red')
```

```python
# Star shaped graph
sg = nx.star_graph(5)
nx.draw(sg,with_labels=True)
```

```python
# Square lattice
N = 5
nxn = nx.grid_2d_graph(N,N)
nx.draw(nxn,with_labels=True)
```

```python
# Plot the networks
ER = nx.erdos_renyi_graph(50,0.2,seed=19)
nx.draw(ER,with_labels=True,node_color='yellow')

#BA = nx.barabasi_albert_graph(50,2,seed=19)
#nx.draw(BA,with_labels=True,node_color='red')
```

```python
# Plot the degree distribution
degs = list(dict(ER.degree()).values())
plt.hist(degs)
```
